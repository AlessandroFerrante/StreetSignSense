## Training yolo12m model

```python
model = YOLO('yolo12m.pt')

results = model.train(
    data='data.yaml',
    imgsz=640,
    epochs=600,
    patience=100,
    save_period=50,
    name='yolo12m_run',
    project='streetsignsense',
    device='0,1',
    batch=40,
    fliplr=0.0
)
```
